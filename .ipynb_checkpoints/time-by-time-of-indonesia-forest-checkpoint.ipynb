```python
import folium

# Define the center of our map.
lat, lon = 6.1750, 106.8283

my_map = folium.Map(location=[lat, lon], zoom_start=10)
my_map
```

```python
def add_ee_layer(self, ee_image_object, vis_params, name):
    """Adds a method for displaying Earth Engine image tiles to folium map."""
    map_id_dict = ee.Image(ee_image_object).getMapId(vis_params)
    folium.raster_layers.TileLayer(
        tiles=map_id_dict['tile_fetcher'].url_format,
        attr='Map Data &copy; <a href="https://earthengine.google.com/">Google Earth Engine</a>',
        name=name,
        overlay=True,
        control=True
    ).add_to(self)

# Add Earth Engine drawing method to folium.
folium.Map.add_ee_layer = add_ee_layer
```

```python
# Select a specific band and dates for land cover.
# lc_img = lc.select('treecover2000').filterDate(i_date).first()
# lc_img = lc.filter(ee.Filter.date('2018-01-01', '2018-05-01')).first().select('FireMask')
# lc_img = lc.select('LC_Type1').filterDate(i_date).first()
# lc_img = lc.select('O3_column_number_density').filterDate('2019-06-01', '2019-06-05')
# lc_img = ee.ImageCollection('COPERNICUS/S5P/NRTI/L3_O3').select('O3_column_number_density').filterDate('2019-06-01', '2019-06-05')
lc_img = ee.ImageCollection('MODIS/061/MOD11A1').filter(ee.Filter.date('2023-01-01', '2023-05-01')).first().select('LST_Day_1km')

# Set visualization parameters for land cover.
lc_vis_params = {
    'min': 13000.0,
  'max': 16500.0,
  'palette': [
    '040274', '040281', '0502a3', '0502b8', '0502ce', '0502e6',
    '0602ff', '235cb1', '307ef3', '269db1', '30c8e2', '32d3ef',
    '3be285', '3ff38f', '86e26f', '3ae237', 'b5e22e', 'd6e21f',
    'fff705', 'ffd611', 'ffb613', 'ff8b13', 'ff6e08', 'ff500d',
    'ff0000', 'de0101', 'c21301', 'a71001', '911003'
  ]
}

# Create a map.
lat, lon = -5.135399, 119.423790
my_map = folium.Map(location=[lat, lon], zoom_start=4.5)

# Add the land cover to the map object.
my_map.add_ee_layer(lc_img, lc_vis_params, 'Land Surface Temperature')


# Add a layer control panel to the map.
my_map.add_child(folium.LayerControl())

# Display the map.
display(my_map)
```
